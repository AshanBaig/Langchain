# 1_Text Loader

- Text loader srf .txt wli file load krta hy



```python
from langchain_community.document_loaders import TextLoader,PyPDFLoader
```

```python
loader = TextLoader("cricket.txt")
docs = loader.load()
print(docs)
```

```python
loader = PyPDFLoader("books/pdf_file.pdf")
docs = loader.load()
print(docs)
```

```python
from langchain_community.document_loaders import DirectoryLoader
```

```python
loader =  DirectoryLoader(
    path = "books",
    glob = "*.pdf",
    loader_cls=PyPDFLoader
)
```

```python
docs = loader.load()  #  bht time le rha load hone me bcz book hy 1  bht bari
print(len(docs))  # 791 is the output means itne pages hain
```

# lazy = jo jo load hota rhyga print krta rhya 

```python
docs = loader.lazy_load()
for i in docs:
    print(i.metadata)
```
